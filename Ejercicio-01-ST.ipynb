##### Para poder resolver los ejercicios mencionados con anterioridad, se debe utilizar la siguiente f√≥rmula:
**Serie de Taylor:**   $T(x)=f(x_{0})+f'(x_{0})(x-x_{0})+f''(x_{0})(x-x_{0})^{2}+\dots+f^{(n)}(x_{0})^{n}+R_{n}$  
Donde $R_{n}$ es el error de truncamiento y representa $R_{n}=\dfrac{f^{(n+1)}(x_{0})}{(n+1)!}(x-x_{0})^{n+1}$ con n el orden de la Serie de Taylor.

```python
import math
import sympy as sym

# Inicializaci√≥n de variables globales
polinomio_taylor = 0

# Calcula la serie de Taylor para una funci√≥n dada hasta un orden m√°ximo
def calcular_serie_taylor(funcion, punto_central, orden_max):
    global polinomio_taylor
    polinomio_taylor = 0
    for i in range(orden_max + 1):
        derivada = funcion.diff(x, i)
        polinomio_taylor += derivada.subs(x, punto_central) * (x - punto_central) ** i / math.factorial(i)

# Imprime la serie de Taylor resultante
def imprimir_serie_taylor():
    global polinomio_taylor
    print("\nSerie de Taylor resultante:")
    print(polinomio_taylor)

# Encuentra el polinomio de Taylor de mejor aproximaci√≥n
def encontrar_serie_taylor(funcion, orden_max, punto_central):
    calcular_serie_taylor(funcion, punto_central, orden_max)
    imprimir_serie_taylor()

# Definici√≥n de la funci√≥n simb√≥lica
x = sym.Symbol('x')
funcion_taylor = 1 / (25 * x ** 2 + 1)

# Par√°metros de evaluaci√≥n
punto_central = 0
orden_max = 5

# Llamada a la funci√≥n para calcular y mostrar la serie de Taylor
encontrar_serie_taylor(funcion_taylor, orden_max, punto_central)
```

##### GR√ÅFICA DE TAYLOR: 

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definir la variable simb√≥lica
x = sym.Symbol('x')

# Definir la funci√≥n
fx1 = 1 / (25 * x ** 2 + 1)

# Calcular la serie de Taylor de orden 5 alrededor de 0
polinomio_taylor_orden_5 = fx1.series(x, 0, 6).removeO()

# Convertir las funciones simb√≥licas a funciones num√©ricas para graficar
fx1_numerica = sym.lambdify(x, fx1, 'numpy')
polinomio_taylor_numerico_orden_5 = sym.lambdify(x, polinomio_taylor_orden_5, 'numpy')

# Rango de valores para x
x_vals = np.linspace(-2, 2, 400)

# Evaluar las funciones
y_fx1 = fx1_numerica(x_vals)
y_serie_taylor_orden_5 = polinomio_taylor_numerico_orden_5(x_vals)

# Graficar la funci√≥n original y la serie de Taylor de orden 5
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_fx1, label='Funci√≥n Original: $f(x) = \\frac{1}{25x^2 + 1}$', color='lightblue')
plt.plot(x_vals, y_serie_taylor_orden_5, label='Serie de Taylor (Orden 5)', color='green', linestyle='--')

# A√±adir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-0.5, 0.5)
plt.ylim(-0.5, 1.50)
plt.title('Comparaci√≥n entre la funci√≥n original y su serie de Taylor de orden 5')
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.show()
```

##### POLINOMIO DE LAGRANGE

```python
import sympy as sym
# Definicion de funciones
# Funci√≥n 1
x = sym.Symbol('x')
fx1 = 1 / (25 * x ** 2 + 1)
```

```python
import sympy as sym

# Definici√≥n de la variable simb√≥lica
x = sym.Symbol('x')

# Funci√≥n 1: f(x) = 1 / (25 * x^2 + 1)
fx1 = 1 / (25 * x ** 2 + 1)

# Puntos para el polinomio de Lagrange
puntos_x = [-0.3, 0, 0.3]
puntos_y = [fx1.subs(x, punto) for punto in puntos_x]  # Evaluaci√≥n de f(x) en los puntos

# Funci√≥n para calcular el polinomio de Lagrange
def lagrange_approx(xs, ys):
    lagrange = 0
    for i in range(len(xs)):
        term = ys[i]
        for j in range(len(xs)):
            if i != j:
                term *= (x - xs[j]) / (xs[i] - xs[j])
        lagrange += term
    return sym.expand(lagrange)

# Calcular el polinomio de Lagrange para los puntos dados
polinomio_lagrange = lagrange_approx(puntos_x, puntos_y)

# Mostrar el polinomio de Lagrange
print("Polinomio de Lagrange:")
print(polinomio_lagrange)
```

##### GR√ÅFICA DE LAGRANGE:

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de la variable simb√≥lica
x = sym.Symbol('x')

# Funci√≥n 1: f(x) = 1 / (25 * x^2 + 1)
fx1 = 1 / (25 * x ** 2 + 1)

# Puntos para el polinomio de Lagrange
puntos_x = [-0.3, 0, 0.3]
puntos_y = [fx1.subs(x, punto) for punto in puntos_x]  # Evaluaci√≥n de f(x) en los puntos

# Funci√≥n para calcular el polinomio de Lagrange
def lagrange_approx(xs, ys):
    lagrange = 0
    for i in range(len(xs)):
        term = ys[i]
        for j in range(len(xs)):
            if i != j:
                term *= (x - xs[j]) / (xs[i] - xs[j])
        lagrange += term
    return sym.expand(lagrange)

# Calcular el polinomio de Lagrange para los puntos dados
polinomio_lagrange = lagrange_approx(puntos_x, puntos_y)

# Convertir las funciones simb√≥licas a funciones num√©ricas para graficar
fx1_func = sym.lambdify(x, fx1, 'numpy')
polinomio_lagrange_func = sym.lambdify(x, polinomio_lagrange, 'numpy')

# Rango de valores para x
x_vals = np.linspace(-1, 1, 400)

# Evaluar las funciones
y_funcion_original = fx1_func(x_vals)
y_polinomio_lagrange = polinomio_lagrange_func(x_vals)

# Graficar la funci√≥n original y el polinomio de Lagrange
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_funcion_original, label='Funci√≥n Original: f(x) = 1 / (25x^2 + 1)', color='orange')
plt.plot(x_vals, y_polinomio_lagrange, label='Polinomio de Lagrange', color='purple', linestyle='--')

# A√±adir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-0.5 ,0.5)
plt.ylim(-0.5,1.50)
plt.title('Comparaci√≥n entre la funci√≥n original y su polinomio de Lagrange')
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.show()


```
