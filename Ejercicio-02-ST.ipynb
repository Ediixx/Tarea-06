```python
import math
import sympy as sym

# Inicialización de variables globales
polinomio_taylor = 0

# Calcula la serie de Taylor para una función dada hasta un orden máximo
def calcular_serie_taylor(funcion, punto_central, orden_max):
    global polinomio_taylor
    polinomio_taylor = 0
    for i in range(orden_max + 1):
        derivada = funcion.diff(x, i)
        polinomio_taylor += derivada.subs(x, punto_central) * (x - punto_central) ** i / math.factorial(i)

# Imprime la serie de Taylor resultante
def imprimir_serie_taylor():
    global polinomio_taylor
    print("\nSerie de Taylor resultante:")
    print(polinomio_taylor)

# Encuentra el polinomio de Taylor de mejor aproximación
def encontrar_serie_taylor(funcion, orden_max, punto_central):
    calcular_serie_taylor(funcion, punto_central, orden_max)
    imprimir_serie_taylor()

# Definición de la función simbólica (arcotangente)
x = sym.Symbol('x')
funcion_taylor = sym.atan(x)  # Usamos arcotangente en lugar de la función anterior

# Parámetros de evaluación
punto_central = 0  # Punto alrededor del cual se expande la serie
orden_max = 3  # Orden máximo de la serie

# Llamada a la función para calcular y mostrar la serie de Taylor
encontrar_serie_taylor(funcion_taylor, orden_max, punto_central)
```

##### GRÁFICA DE TAYLOR: 

```python
import math
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definir la variable simbólica
x = sym.Symbol('x')

# Definir la función original (arcotangente de x)
funcion_original = sym.atan(x)

# Definir el polinomio de Taylor de orden 3 alrededor de 0
polinomio_taylor = funcion_original.series(x, 0, 4).removeO()

# Convertir las funciones simbólicas a funciones numéricas para graficar
funcion_original_numerica = sym.lambdify(x, funcion_original, 'numpy')
polinomio_taylor_numerico = sym.lambdify(x, polinomio_taylor, 'numpy')

# Rango de valores para x
x_vals = np.linspace(-2, 2, 400)

# Evaluar las funciones
y_funcion_original = funcion_original_numerica(x_vals)
y_serie_taylor = polinomio_taylor_numerico(x_vals)

# Graficar la función original y la serie de Taylor
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_funcion_original, label='Función Original: arctan(x)', color='blue')
plt.plot(x_vals, y_serie_taylor, label='Serie de Taylor (Orden 3)', color='red', linestyle='--')

# Añadir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación entre la función original y su serie de Taylor')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

```

##### POLINOMIO DE LAGRANGE

```python
import sympy as sym

# Definición de la variable simbólica
x = sym.Symbol('x')

# Función 1: Arcotangente de x
fx1 = sym.atan(x)

# Puntos para el polinomio de Lagrange
puntos_x = [-2, 0, 2]
puntos_y = [fx1.subs(x, punto) for punto in puntos_x]  # Evaluación de f(x) en los puntos

# Función para calcular el polinomio de Lagrange
def lagrange_approx(xs, ys):
    lagrange = 0
    for i in range(len(xs)):
        term = ys[i]
        for j in range(len(xs)):
            if i != j:
                term *= (x - xs[j]) / (xs[i] - xs[j])
        lagrange += term
    return sym.expand(lagrange)

# Calcular el polinomio de Lagrange para los puntos dados
polinomio_lagrange = lagrange_approx(puntos_x, puntos_y)

# Mostrar el polinomio de Lagrange
print("Polinomio de Lagrange:")
print(polinomio_lagrange)

```

##### GRÁFICA DE LAGRANGE:

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definición de la variable simbólica
x = sym.Symbol('x')

# Función 1: Arcotangente de x
fx1 = sym.atan(x)

# Puntos para el polinomio de Lagrange
puntos_x = [-2, 0, 2]
puntos_y = [fx1.subs(x, punto) for punto in puntos_x]  # Evaluación de f(x) en los puntos

# Función para calcular el polinomio de Lagrange
def lagrange_approx(xs, ys):
    lagrange = 0
    for i in range(len(xs)):
        term = ys[i]
        for j in range(len(xs)):
            if i != j:
                term *= (x - xs[j]) / (xs[i] - xs[j])
        lagrange += term
    return sym.expand(lagrange)

# Calcular el polinomio de Lagrange para los puntos dados
polinomio_lagrange = lagrange_approx(puntos_x, puntos_y)

# Convertir las funciones simbólicas a funciones numéricas para graficar
fx1_func = sym.lambdify(x, fx1, 'numpy')
polinomio_lagrange_func = sym.lambdify(x, polinomio_lagrange, 'numpy')

# Rango de valores para x
x_vals = np.linspace(-3, 3, 400)

# Evaluar las funciones
y_funcion_original = fx1_func(x_vals)
y_polinomio_lagrange = polinomio_lagrange_func(x_vals)

# Graficar la función original y el polinomio de Lagrange
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_funcion_original, label='Función Original: arctan(x)', color='green')
plt.plot(x_vals, y_polinomio_lagrange, label='Polinomio de Lagrange', color='yellow', linestyle='--')

# Añadir etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación entre la función original y su polinomio de Lagrange')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

```
